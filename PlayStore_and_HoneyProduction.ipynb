## 6. Check if there are missing values present in the column Rating, If any? drop them and and create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

```python
data['Rating'].isnull().sum()
```

```python
data=data.dropna()
```

```python
data['Rating'].isnull().sum()
```

```python
# We have created the new column and applied lambda for high and low rating.
data['Rating_category']=data['Rating'].apply(lambda x: 'High' if x>3.5 else 'Low')
```

## 8. Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation

```python
# We have used replace and applied eval function for converting it to numeric.
data['Reviews']=data['Reviews'].str.replace("M","*1000000").apply(eval)
```

```python
# Now the datatype is changed to int64
data['Reviews']
```

```python
# We used boxplot to see a outliers
sns.boxplot(data['Reviews'])
plt.show()
```

```python
data['Reviews']=np.log(data['Reviews'])
data['Reviews']
```

```python
# Using log function we have removed the outliers
sns.boxplot(data['Reviews'])
plt.show()
```

## 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')

```python
data['Size'].unique()
```

```python
# We have dropped the Varies with device
data=data[data['Size']!='Varies with device']
```

```python
data['Size']=data['Size'].str.replace("k","*1000").str.replace("M","*1000000").apply(eval)
```

```python
data['Size']
```

## 10. Check the column 'Installs', treat the unwanted characters and convert the column into a suitable data type

```python
data['Installs'].unique()
```

```python
data['Installs']=data['Installs'].str.replace(",","").str.replace("+","").astype(int)
```

```python
data['Installs'].unique()
```

## 11. Check the column 'Price' , remove the unwanted characters and convert the column into a suitable data type

```python
data['Price'].unique()
```

## 12. Drop the columns which you think redundant for the analysis.(suggestion: drop column 'rating', since we created a new feature from it (i.e. rating_category) and the columns 'App', 'Rating' ,'Genres','Last Updated', 'Current Ver','Android Ver' columns since which are redundant for our analysis)

```python
# We have dropped the redundant columns
data=data.drop(['App','Rating' ,'Genres','Last Updated', 'Current Ver','Android Ver'],axis=1)
```

## 13. Encode the categorical columns.


```python
# Using LabelEncoder we have encoded the 'Rating_category'
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Rating_category']=le.fit_transform(data['Rating_category'])
```

```python
#Using get_dummies we have encoded the categorical columns.
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

## 14. Segregate the target and independent features (Hint: Use Rating_category as the target)


```python
# We have assigned the 'Rating_category' as dependent and all other columns as independent.
x=data.drop(['Rating_category'],axis=1)
y=data[['Rating_category']]
```

```python
y
```

## 15. Split the dataset into train and test.


```python
#Using train_test_split we have trained the 70% of data and tested the 30% of data.
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

## 16. Standardize the data, so that the values are within a particular range
